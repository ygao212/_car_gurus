```python
from _clean import data
import matplotlib.pyplot as plt
%matplotlib inline
data["price-market_price"] = data.price - data.market_price
data.head()
```

```python
# On average, cars selling in NJ have higher mileage by 10K miles than those in NY, 
# whereas price and market_price in both states are about the same
bar_data = data.loc[:,("state","mileage","price","market_price","year")].groupby(["state"]).mean()

fig = plt.figure()

ax = fig.add_subplot(111)
ax2 = ax.twinx()

ax.set_ylabel("price and mileage")
ax2.set_ylabel("year")
ax.set_title("Mileage, Year, Price, and Market_Price by State")

width = 0.07

bar_data.mileage.plot(kind="bar", color="#ADD8E6", ax=ax, width=width, position=1)
bar_data.price.plot(kind="bar", color="#90EE90", ax=ax, width=width, position=4)
bar_data.market_price.plot(kind="bar", color="#FFB6C1", ax=ax, width=width, position=3)
bar_data.year.plot(kind="bar", color="#FAFAD2", ax=ax2, width=width, position=2, grid=True)


rects = ax.patches
for rect in rects:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, round(height), ha='center', va='bottom')

ax.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.);

rects = ax2.patches
for rect in rects:
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width()/2, height, round(height), ha='center', va='bottom')

ax2.legend(bbox_to_anchor=(1.1, 0.72), loc=2, borderaxespad=0.);
```

```python
# On average, cars selling in NJ have higher mileage than those in NY, 
# whereas price and market_price are about the same
bar_data = data.loc[:,("state", "days_listed", "dealer_rating")].groupby(["state"]).mean()

fig = plt.figure(facecolor="white")

ax = fig.add_subplot(111)
ax2 = ax.twinx()

width = 0.2

bar_data.days_listed.plot(kind="bar", color="#20B2AA", ax=ax, width=width, position = 0, grid=True)
bar_data.dealer_rating.plot(kind="bar", color="#FFA07A", ax=ax2, width=width, position= 1)

ax.set_title("Dealer_Rating and Days_Listed by State")
ax.set_ylabel("days_listed")
ax.set_ylim(0,90)
ax2.set_ylabel("dealer_rating")
ax2.set_ylim(0,5)

rects = ax.patches
for rect in rects:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height, round(height), ha='center', va='bottom')

ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);

rects = ax2.patches
for rect in rects:
    height = rect.get_height()
    ax2.text(rect.get_x() + rect.get_width()/2, height, round(height), ha='center', va='bottom')

ax2.legend(bbox_to_anchor=(1.05, 0.9), loc=2, borderaxespad=0.);
```
